N.B. tous les calculs sont effectués en posant $abs2(\overrightarrow{\rm V}) = \overrightarrow{\rm V}.\overrightarrow{\rm V}$

## 1. Modélisation de la sphère

### 1.1 Calcul de l'intersection entre la sphère et une droite

Soit une sphère $S$ de centre $\rm C$ et de rayon $R$.

La sphère est constituée des points $\rm P$ tels que (1) :
$$abs2(\overrightarrow{\rm CP}) = R^2$$

Soit une droite passant par $\rm O$, et de vecteur directeur $\overrightarrow{\rm D}$.<br>
Elle est donc
constituée des points $\rm P$ pour lesquels il existe un coefficient $k$ tel que (2) :

$$(\overrightarrow{\rm OP}) = k\,\overrightarrow{\rm D}$$

En développant (1) on obtient (3) :
$${\rm (P_x - C_x)^2  + (P_y - C_y)^2 + (P_z - C_z)^2} = R^2$$

En développant (2) on obtient (4) :
$$
{\rm P_x} = k\,{\rm D_x + O_x} \\
{\rm P_y} = k\,{\rm D_y + O_y} \\
{\rm P_z} = k\,{\rm D_z + O_z}
$$

En injectant (4) dans (3), on obtient l'équation (5) :

$$(k\,{\rm D_x + O_x - C_x})^2 +
  (k\,{\rm D_y + O_y - C_y})^2 +
  (k\,{\rm D_z + O_z - C_z})^2 = R^2
$$

Que l'on peut développer en (6) :

$$\begin{align*}
k^2{\rm D_x}^2 + 2\,k\,{\rm D_x(O_x - C_x)} &+ {\rm (O_x - C_x)^2} + \\
k^2{\rm D_y}^2 + 2\,k\,{\rm D_y(O_y - C_y)} &+ {\rm (O_y - C_y)^2} + \\
k^2{\rm D_z}^2 + 2\,k\,{\rm D_z(O_z - C_z)} &+ {\rm (O_z - C_z)^2} = R^2
\end{align*}$$

En introduisant les produits scalaires dans (6), cela donne (7) :

$$k^2\,\overrightarrow{\rm D}.\overrightarrow{\rm D} + 2\,k\,\overrightarrow{\rm D}.(\overrightarrow{\rm CO}) + (\overrightarrow{\rm CO}).(\overrightarrow{\rm CO}) = R^2$$

Que l'on peut simplifier en notant que $\overrightarrow{\rm D}$ est normé, pour obtenir une équation du second degré en $k$ (8) :

$$k^2 + 2\,k\,\overrightarrow{\rm D}.(\overrightarrow{\rm CO}) + abs2(\overrightarrow{\rm CO}) - R^2 = 0$$

On renvoie la plus petite des solutions positives, sinon $FARAWAY$.

### 1.2 Calcul de la normale en un point de la sphère

La normale en un point $\rm M$ de la sphère est calculée par : 
$$\overrightarrow{\rm N} = \frac{\overrightarrow{\rm CM}}{R}$$


## 2. Modélisation d'un plan

### 2.1 Intersection avec une droite

Soit un plan défini par un point $\rm P$ et sa normale $\overrightarrow{\rm N}$.

L'ensemble des points $\rm M$ du plan vérifie (1) : $\overrightarrow{\rm PM}.\overrightarrow{\rm N} = 0$

Soit la droite passant par $\rm O$ de vecteur directeur $\overrightarrow{\rm D}$.

Les points M de la droite vérifient (2) : ${\rm M} = {\rm O} + k\,\overrightarrow{\rm D}$

Il existe un point d'intersection, sauf lorsque $\overrightarrow{\rm D}$ est parallèle au plan.<br>
La condition d'existence d'un point d'intersection est (3) : $\overrightarrow{\rm D}.\overrightarrow{\rm N} \neq 0$ 

En injectant (2) dans (1) :
$$\overrightarrow{({\rm O} + k\,\overrightarrow{\rm D} - {\rm P})}.\overrightarrow{\rm N} = 0$$

Ce qui peut encore s'écrire :
$$ k\,\overrightarrow{\rm D}.\overrightarrow{\rm N} = \overrightarrow{\rm (P - O)}.\overrightarrow{\rm N} = \overrightarrow{\rm PO}.\overrightarrow{\rm N} $$

D'où on déduit (si la condition (3) est vérifiée) :
$$k = \frac{\overrightarrow{\rm OP}.\overrightarrow{\rm N}}{\overrightarrow{\rm D}.\overrightarrow{\rm N}}$$

Que l'on renvoie si $k > 0$.

### 2.2 Calcul de la normale.

On renvoie $\overrightarrow{\rm N}$, donnée pour la définition du plan.

## 3. Modélisation d'un polygone
### 3.1 Intersection avec une droite

Soit un polygone $P$ donné par :
- une liste de points donnés par leurs coordonnées 2D dans un repère orthonormé $(\overrightarrow{P_u},\overrightarrow{P_v})$ lié au polygone,
- les coordonnées 3D des vecteurs du repère $(\overrightarrow{P_u},\overrightarrow{P_v})$
- les coordonnées 3D de l'origine du repère $P_o$

On alors : $P = \{\;P_i\; /\; P_i = P_o + p_{iu} \overrightarrow{P_u} + p_{iv} \overrightarrow{P_v}\;\}\;$
<!--
&nbsp;où :
$\bigg\{\begin{align*}
P_u &= P_{ux} X + P_{uy} Y + P_{uz} Z \\
P_v &= P_{vx} X + P_{vy} Y + P_{vz} Z
\end{align*}$
-->

Le polygone définit un plan, passant par le point $P_o$ et de normale $\overrightarrow{P_u} \times \overrightarrow{P_v}$

Pour chercher l'intersection d'un rayon avec le polygone, on détermine d'abord les coordonnées 3D de l'intersection
$\rm M$ du rayon avec le plan du polygone, de la même manière que pour le plan infini :
$$ {\rm M} = \begin{bmatrix} m_x \\ m_y \\ mz \end{bmatrix}$$
et on cherche ses coordonnées $(m_u,mv)$ dans le repère du polygone.

En développant on obtient : ${\rm M} = P_o + m_u \overrightarrow{P_u} + m_v \overrightarrow{P_v} $

qui donne : $\overrightarrow{{\rm M} - P_o} = m_u \overrightarrow{P_u} + m_v \overrightarrow{P_v}$

Comme nous savons par ailleurs que $(\overrightarrow{P_u},\overrightarrow{P_v})$ est orthonormé, donc :

<div style="text-align: center; margin-top: 1em;">
$\overrightarrow{P_u} . \overrightarrow{P_u} = 1$
&nbsp; et &nbsp;
$\overrightarrow{P_v} . \overrightarrow{P_v} = 1
$&nbsp; et &nbsp;
$\overrightarrow{P_u} . \overrightarrow{P_v} = 0$</div>

nous pouvons en déduire que :

$$\Bigg\{\begin{align*}
m_u &=\overrightarrow{P_o\rm M}.\overrightarrow{P_u} \\
m_v &= \overrightarrow{P_o \rm M}.\overrightarrow{P_v}
\end{align*}$$


A partir de là il faut déterminer si le point $\rm M$ fait partie ou non du polygone.
Pour cela on lance un rayon semi-infini partant de $\rm M$ dans le sens de $P_u$ (i.e. $v$ fixé, $u$ croissant) et on compte le nombre d'intersections avec les segments du polygone. Si le nombre est pair, $\rm M$ est en-dehors du polygone, s'il est impair $\rm M$ est dedans.

Pour chacun des segments $[P_i, P_{i+1}]$ du polygone:
- si $P_{iv}$ est égal à $P_{(i+1)v}$ (segment parallèle au rayon) on considère qu'il n'y a pas intersection,
- si $m_v \not\in [min(P_{iv},P_{(i+1)v}), max(P_{iv},P_{(i+1)v})]$ (rayon au-dessus ou en-dessous du segment) il n'y a pas intersection, 
- sinon, soit $v = a\,u + b$ l'équation de la droite portant le segment. On a :

$$\begin{align*}
v_i &= a\,u_i + b \\
v_{i+1} &= a\,u_{i+1} + b
\end{align*}$$<br>
D'où on déduit : $b = v_i - a\,u_i$&nbsp;
puis &nbsp;$a = \dfrac{v_{i+1} - v_i}{u_{i+1} - u_i}$&nbsp;
et $b = \dfrac{u_{i+1} v_i - u_i v_{i+1}}{u_{i+1} - u_i}$<br>
L'intersection avec la droite horizontale $v = m_v$ passant par $\rm M$ se trouve en $u = \frac{m_v - b}{a}$, et elle nous intéresse si elle se trouve du côté positif de $\rm M$ :

$$u - m_u > 0$$
On compte donc $+1$, si :
$$\dfrac{m_v - b}{a} \;=\; m_v \dfrac{u_{i+1} - u_i}{v_{i+1} - v_i} - \dfrac{u_{i+1} v_i - u_i v_{i+1}}{v_{i+1} - v_i} \; > \;m_u
$$

N.B. on peut précalculer les coefficients de cette équation, ainsi que $v_{min}$ et $v_{max}$ pour chacun des segments.

### 3.2 Calcul de la normale

Dans la mesure où le point se trouve au sein du polygone la normale est celle du plan, soit : $P_u \times P_v$

### 3.3 Cube

Soit un cube défini par son centre $\rm C$, le repère orthonormé $(\overrightarrow{P_u},\overrightarrow{P_v})$ des arêtes de l'une de ses faces, et la longueur $l$ de ses arêtes.

Le cube sera modélisé par un ensemble de 6 polygones
défini chacun par point de référence correspondant à l'un des coins du cube,
un repère correspondant à deux arêtes, et 4 sommets dont les coordonnées sont :
$$
P_1i = \{(0,0),(0,1),(1,1),(1,0)\}
$$

En posant $\overrightarrow{P_w} = \overrightarrow{P_u} \times \overrightarrow{P_v}$,
les sommet, repère, et sens de la normale des six polygones sont :

$$
{\rm P_{1o}} = {\rm C} - \frac{l}{2}\,\overrightarrow{P_u} - \frac{l}{2}\,\overrightarrow{P_v} - \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{1u}},\overrightarrow{P_{1v}}) = (\overrightarrow{P_u},\overrightarrow{P_v}) \qquad
ns_1 = -1
\\
{\rm P_{2o}} = {\rm C} - \frac{l}{2}\,\overrightarrow{P_u} - \frac{l}{2}\,\overrightarrow{P_v} + \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{2u}},\overrightarrow{P_{2v}}) = (\overrightarrow{P_u},\overrightarrow{P_v}) \qquad
ns_2 = +1
\\
{\rm P_{3o}} = {\rm C} - \frac{l}{2}\,\overrightarrow{P_u} - \frac{l}{2}\,\overrightarrow{P_v} - \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{3u}},\overrightarrow{P_{3v}}) = (\overrightarrow{P_w},\overrightarrow{P_v}) \qquad
ns_3 = +1
\\
{\rm P_{4o}} = {\rm C} + \frac{l}{2}\,\overrightarrow{P_u} - \frac{l}{2}\,\overrightarrow{P_v} - \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{4u}},\overrightarrow{P_{4v}}) = (\overrightarrow{P_w},\overrightarrow{P_v}) \qquad
ns_4 = -1
\\
{\rm P_{5o}} = {\rm C} - \frac{l}{2}\,\overrightarrow{P_u} - \frac{l}{2}\,\overrightarrow{P_v} - \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{5u}},\overrightarrow{P_{5v}}) = (\overrightarrow{P_u},\overrightarrow{P_w}) \qquad
ns_5 = +1
\\
{\rm P_{6o}} = {\rm C} - \frac{l}{2}\,\overrightarrow{P_u} + \frac{l}{2}\,\overrightarrow{P_v} - \frac{l}{2}\,\overrightarrow{P_w} \qquad
(\overrightarrow{P_{6u}},\overrightarrow{P_{6v}}) = (\overrightarrow{P_u},\overrightarrow{P_w}) \qquad

## 4. Modélisation d'un disque

Le modèle du disque comprend :
- son centre $\rm C$,
- sa normale $\overrightarrow{\rm N}$,
- son rayon $R$

Pour calculer le point d'intersection avec une droite
on procède d'abord comme pour le plan, puis on calcule la distance de $\rm M$ à $\rm C$ :

$$ d = abs(CM)$$

Le point $\rm M$ fait partie du disque si $d < R$.

La normale au disque est triviale puisqu'il s'agit de $\overrightarrow{\rm R}$.

## 5. Modélisation d'un cylindre

### 5.1 Point d'intersection avec une droite

Soit un cylindre donnée par :
- un point $\rm P$ de son axe,
- $\overrightarrow{\rm V}$ vecteur directeur de son axe,  que l'on suppose normé, 
- son rayon $R$

Les points $\rm M$ du cylindre se trouvent à la distance $R$ de l'axe, ce que l'on exprime par (1) :

$$\frac{abs2(\overrightarrow{\rm PM} \times \overrightarrow{\rm V})}{abs2(\overrightarrow{\rm V})} = R^2$$

Comme on suppose que $\overrightarrow{\rm V}$ est normé, $abs2(\overrightarrow{\rm V}) = 1$, donc :

$$abs2(\overrightarrow{\rm PM} \times \overrightarrow{\rm V}) = R^2$$

Par ailleurs les points $\rm M$ de la droite de vecteur directeur $\rm D$ passant par $\rm O$ vérifient comme d'habitude (2):

$$\overrightarrow{\rm OM} = k\,\overrightarrow{\rm D}$$

En notant dans (1) que $\overrightarrow{\rm PM} = \overrightarrow{\rm PO} + \overrightarrow{\rm OM}$ et en injectant (2) :

$$\begin{align*}
abs2((\overrightarrow{\rm PO} + k\,\overrightarrow{\rm D}) \times \overrightarrow{\rm V}) &= R^2 \\
abs2(\overrightarrow{\rm PO} \times \overrightarrow{\rm V} + k\,\overrightarrow{\rm D} \times \overrightarrow{\rm V}) &=R^2 \\
\end{align*}$$

Posons $\overrightarrow{\rm W} = \overrightarrow{\rm PO} \times \overrightarrow{\rm V}$
et $\overrightarrow{\rm T} = \overrightarrow{\rm D} \times \overrightarrow{\rm V}$, alors :

$$abs2(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}) = R^2$$

En notant que $abs2(\overrightarrow{\rm X}) = \overrightarrow{\rm X}.\overrightarrow{\rm X}$ :

$$(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}).(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}) = R^2$$

ou encore :

$$k^2\,\overrightarrow{\rm T}.\overrightarrow{\rm T} + 2\,k\,\overrightarrow{\rm T}.\overrightarrow{\rm W} + \overrightarrow{\rm W}.\overrightarrow{\rm W} = R^2$$

qu'il va falloir résoudre en k.

### 5.2 Calcul de la normale

<u>Solution 1 :</u>

La normale en un point $\rm M$ du cylindre peut se calculer en normant le vecteur de longueur $R$ :
<span style="color:#C00">_(TODO : Un schéma serait bienvenu)_</span>

$$\overrightarrow{\rm N} = (\overrightarrow{\rm MP} \times \overrightarrow{\rm V} \times \overrightarrow{\rm V})_{normé}$$


En effet, $\overrightarrow{\rm PM} \times \overrightarrow{\rm V}$ est un vecteur de longueur $R$,
perpendiculaire à $\overrightarrow{\rm V}$ (ce qui est bien),
mais aussi à $\overrightarrow{\rm PM}$ (ce qui est moins bien).
Pour le ramener dans la bonne direction, on refait un produit vectoriel avec $\overrightarrow{\rm V}$,
et on change de signe.

<u>Solution 2 :</u>

Une deuxième solution peut-être plus intuitive, consiste à normer la composante radiale obtenue en retranchant à $\overrightarrow{\rm PM}$ sa composante axiale :

$$\overrightarrow{\rm N} = (\overrightarrow{\rm PM} - \overrightarrow{\rm PM}.\overrightarrow{\rm V}\,\overrightarrow{\rm V})_{normé}$$

Cette méthode est légèrement moins rapide que la précédente, car elle nécessite une opération de plus et impose de stocker $\overrightarrow{\rm PM}$ dans une variable.

### 5.3 Cylindre tronqué

Le cylindre tronqué se modélise comme un cylindre infini en précisant une hauteur, et en considérant que le point $\rm P$ sur son axe permet également de positionner sa base. On a donc :
- un point $\rm P$ de son axe, donnant également la position de la base du cylindre,
- $\overrightarrow{\rm V}$ le vecteur directeur de son axe,  que l'on suppose normé, 
- son rayon $R$
- sa hauteur h.

Pour calculer le point d'intersection d'un cylindre tronqué avec une droite on procède comme pour le cylindre infini, puis on calcule la hauteur du point $\rm M$ au-dessus de la base et le long de l'axe du cylindre : 
$$ h_M = \overrightarrow{\rm PM}.\overrightarrow{\rm V}$$

Le point $\rm M$ est sur le cylindre si $h_M > 0$ et $h_M < h$.

La normale se calcule ensuite comme pour le cylindre infini.

Pour munir le cylindre tronqué de ses extrémités, il faut lui adjoindre les deux disques :

## 6. Modélisation d'un cône

### 6.1 Point d'intersection avec une droite

Soit le cône donné par :
- son sommet $\rm P$
- $\overrightarrow{\rm V}$ vecteur directeur de son axe,  que l'on suppose normé, 
- son coefficient $\lambda$ tel que le rayon du cône en $\rm M$ soit égal à
$R = \lambda\,\overrightarrow{\rm PM}.\overrightarrow{\rm V}$

Les points $\rm M$ du cône obéissent à l'équation (1) :

$$abs2(\overrightarrow{\rm PM} \times \overrightarrow{\rm V}) =
\lambda^2\,(\overrightarrow{\rm PM}.\overrightarrow{\rm V})^2$$

Par ailleurs les points $\rm M$ de la droite de vecteur directeur $\rm D$ passant par $\rm O$ vérifient comme d'habitude (2):

$$\overrightarrow{\rm OM} = k\,\overrightarrow{\rm D}$$

En notant dans (1) que $\overrightarrow{\rm PM} = \overrightarrow{\rm PO} + \overrightarrow{\rm OM}$ et en injectant (2) :

$$
abs2((\overrightarrow{\rm PO} + k\,\overrightarrow{\rm D}) \times \overrightarrow{\rm V}) =
\lambda^2\,((\overrightarrow{\rm PO} + k\,\overrightarrow{\rm D}).\overrightarrow{\rm V})^2
$$

Soit :
$$
abs2(\overrightarrow{\rm PO} \times \overrightarrow{\rm V} + k\,\overrightarrow{\rm D} \times \overrightarrow{\rm V}) =
\lambda^2\,(\overrightarrow{\rm PO}.\overrightarrow{\rm V} + k\,\overrightarrow{\rm D}.\overrightarrow{\rm V})^2
$$

Posons $\overrightarrow{\rm W} = \overrightarrow{\rm PO} \times \overrightarrow{\rm V}$,
$\quad \overrightarrow{\rm T} = \overrightarrow{\rm D} \times \overrightarrow{\rm V}$,
$\quad w = \overrightarrow{\rm PO} . \overrightarrow{\rm V}$ et
$\quad t = \overrightarrow{\rm D} . \overrightarrow{\rm V}$, alors :

$$abs2(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}) = \lambda^2\,(w + k\,t)^2$$

En notant que $abs2(\overrightarrow{\rm X}) = \overrightarrow{\rm X}.\overrightarrow{\rm X}$ :

$$(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}).(\overrightarrow{\rm W} + k\,\overrightarrow{\rm T}) =
\lambda^2\,(w + k\,t)^2$$

ou encore :

$$
k^2\,(\overrightarrow{\rm T}.\overrightarrow{\rm T} - \lambda^2\,t^2) +
2\,k\,(\overrightarrow{\rm T}.\overrightarrow{\rm W} - \lambda^2\,w\,t) +
(\overrightarrow{\rm W}.\overrightarrow{\rm W} - \lambda^2\,w^2)= 0$$

qu'il va falloir résoudre en k.

### 6.2 Calcul de la normale

On calculera la normale en un point $\rm M$ situé sur la partie supérieure (positive) du cône en normant le vecteur :
<span style="color:#C00">_( TODO : Schéma ! )_</span>

$$\overrightarrow{\rm N} = (\overrightarrow{\rm MP} \times \overrightarrow{\rm V} \times \overrightarrow{\rm PM})_{normé}$$

Sur la partie inférieure il faut inverser le signe :

$$\overrightarrow{\rm N} = (\overrightarrow{\rm MP} \times \overrightarrow{\rm V} \times \overrightarrow{\rm MP})_{normé}$$


### 6.3 Cône tronqué

Le cône tronqué est défini par :
- son sommet $\rm P$
- $\overrightarrow{\rm V}$ vecteur directeur de son axe,  que l'on suppose normé, 
- son coefficient $\lambda$ tel que le rayon du cône en $\rm M$ soit égal à
$R = \lambda\,\overrightarrow{\rm PM}.\overrightarrow{\rm V}$
- l'altitude $h_1$ de sa base, qui peut être positive, négative, ou nulle
- l'altitude de son sommet $h_2$, (positive, négative, ou nulle), mais avec $h_1 < h_2$

On calcule $\rm M$ le point d'intersection avec une droite comme pour le cône infini, puis : 
$$ h_M = \overrightarrow{\rm PM}.\overrightarrow{\rm V}$$

Le point $\rm M$ est sur le cône tronqué si $h_M > h_1$ et $h_M < h_2$.

La normale se calcule ensuite comme pour le cône infini.

Sauf si $h_1$ est nul, la base du cône tronqué correspond au disque :

$$
{\rm C_1} = \overrightarrow{\rm P} + h_1\,\overrightarrow{\rm V} \qquad
\overrightarrow{\rm N_1} = -\overrightarrow{\rm V} \qquad
{\rm R_1} = abs2(k\,h_1)
$$
De même, si $h_2$ est non nul :
$$
{\rm C_2} = \overrightarrow{\rm P} + h_2\,\overrightarrow{\rm V} \qquad
\overrightarrow{\rm N_2} = \overrightarrow{\rm V} \qquad
{\rm R_2} = abs2(k\,h_2)
$$
